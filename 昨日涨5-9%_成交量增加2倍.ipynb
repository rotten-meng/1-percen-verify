交易想法验证

当天涨幅在5%—8%之间 and 当天成交量>昨日成交量的3倍 买入

第二天按买入价 * 1.01卖出，如果不能成交，则在收盘时卖出。

```python
import pandas as pd
import numpy as np
import os
```

```python
def mMain_Fun(mFilepath_str,symbol,mSUMProfit,mTradeNum):
    mData = pd.read_csv(mFilepath_str)
    mLen = len(mData['volume'])
    mClose = np.array(mData['close'])
    mVolume = np.array(mData['volume'])
    mHigh = np.array(mData['high'])
    mLow = np.array(mData['low'])
    mRSI = np.array(mData['RSI'])
    
    mBuyRecord = []
    mSellRecord = []

    mProfit = []


    mHasBuy = False
    
    mTradNum = 0

    i = 1
    while(i < mLen - 10):
        if mHasBuy:
            mHasBuy = False
            mhigh = max(mHigh[i],mHigh[i+1],mHigh[i+2],mHigh[i+3],mHigh[i+4])
            mlow = min(mLow[i],mLow[i+1],mLow[i+2],mLow[i+3],mLow[i+4])
            if mhigh > mClose[i-1] * 1.01:
                if mClose[i-1] * 1.01 > mlow:
                    mProfit.append(0.01)
                else:
                    mProfit.append((mlow - mClose[i-1])/mClose[i-1])
            else:
                mProfit.append((mClose[i+4] - mClose[i-1])/mClose[i-1])

        if mClose[i] > mClose[i-1] * 1.05 and mClose[i] < mClose[i-1] * 1.09 \
        and mVolume[i] > mVolume[i-1] * 3:
        #if mRSI[i] >= 0.6 and mRSI[i-1] < 0.6:
            mHasBuy = True
            mTradNum += 1
        i += 1

    #for i in  mProfit:
        #print(i)

    mTest = np.array(mProfit)
    mSumprofit = mTest.sum()
    print(symbol,'  buy num:',len(mTest),'sum: ',mSumprofit)
    mSUMProfit.append(mSumprofit)
    mTradeNum.append(mTradNum)
```
